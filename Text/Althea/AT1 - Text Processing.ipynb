# String Mixer

```python
str1 = input("Give me a string: ")
str2 = input("Give me another string: ")
str3 = ""
for ch in range(0, len(str1)):
    str3 += str1[ch] + str2[ch]
print(str3)
```

# Palindrome Checker

```python
word = input("Give me a string: ")
word = word.lower()
str2 = ""
x = 0
palindrome = True
for ch in range(0, len(word)):
    if word[ch] != word [x-1]:
        str2 += word[ch] + word [x-1]
        print(word + " is not a palindrome. ")
        palindrome = False
        break
    else:
        x = x - 1
if (palindrome):
    print(word + " is a palindrome. ")
```

or

```python
word = input("Give me a string: ")
clword = word.lower()
clword = clword.replace(" ", "")
clword = ''.join(filter(str.isalnum, clword))
reverse = ""

for ch in clword:
    reverse = ch + reverse
if (clword == reverse):
    print(word, "is a palindrome.")
else:
    print(word, "is not a palindrome.")
```
