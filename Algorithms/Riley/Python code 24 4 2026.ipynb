Battle Ship Game

```python
import random
w = 6
h = 6
ship_count = 4
shots = 15
board = {}
ship_x = random.randint(0, 6)
ship_y = random.randint(0, 6)
board[(ship_x, ship_y)] = "X"


while len(board) != 4:
    ship_x = random.randint(0, 6)
    ship_y = random.randint(0, 6)
    if (ship_x, ship_y) not in board:
        board[(ship_x, ship_y)] = "X"
    else:
        print("Regenerating...")

print(board)




while (shots !=0) and (len(board) != 0):
    guess_x = int(input("guess X"))

    guess_y = int(input("guess Y"))
    shots -= 1
    if (guess_x, guess_y) in board:
        print("hit", guess_x, guess_y, "shots left:", shots)
        del(board[(guess_x, guess_y)])
    else:
        print("miss", (guess_x, guess_y), "shots left:", shots)
    if shots == 0:
        print("lose")
    if (len(board) == 0):
        print("win")
print("ships left:",board)
```

```python
searchlist = [1, 2, 3, 4, 5, 6, 7, 10]
start = 0
end = len(searchlist) -1
found = False
midpos = round(start + end)/2
target = input("find a number")
while (not found) and (end >= start):
    mid = round((start +end)/2)
    print(f"start={start}, end={end}, mid = {mid}, content = {searchlist[midpos]}")
    if searchlist[mid] == target: 
        print(f"target found at position mid")
    if mid > target:
        end = mid -1
    if mid < target:
        start = mid + 1
    if mid == start == end != target:
        break
```
