```python
def bsearch(searchlist, target):
    start = 0
    end = len(searchlist) - 1

    while start <= end:
        mid = (start + end) // 2

        if searchlist[mid] == target:
            print(f"Target found at position {mid}")
            return mid
        elif searchlist[mid] < target:
            start = mid + 1
        else:
            end = mid - 1

    print("Target not found")
    return None

mylist = [1,2,3,5,6,7,100,240,6767,8788,55555,55678]
print(bsearch(mylist, 55678))
```

```python
import random
import math

def distance(px,py):
    return math.sqrt(px*px + py*py)
def MonteCarlo(num_points):
    inQuadrant = 0
    for i in range(num_points):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if distance(x,y) <=1:
            if (y > x - 0.5):
                if (y < x + 0.5):
                    inQuadrant += 1
    return (inQuadrant / num_points) * 4


print(MonteCarlo(1000))
print(MonteCarlo(10000000))


```
