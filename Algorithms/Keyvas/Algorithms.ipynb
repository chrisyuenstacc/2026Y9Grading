# Algorithm

```python
listb = [1,3,5,8,9,15,17,35,47,58,69,75,94,102,777,100000007032]
def bsearch(slist, target):
    startpos=0
    endpos = len(slist) - 1
    found = False
    while (not found) and (endpos>=startpos):
        midpos = round((startpos + endpos)/2)
        if slist[midpos] == target:
            found = True
            print(f"found at {midpos}")
        elif (slist[midpos] < target):
            startpos = midpos + 1
        elif (slist[midpos] > target):
            endpos = midpos - 1
        print(f"start = {startpos} end = {endpos} mid = {midpos} content={slist[midpos]}")
    if found:
        return midpos
    else:
        return -1
print(bsearch(listb,67))
print(bsearch(listb,58))
```

# Monte Carlo Method Pi Estimation

```python
import random
import math
import matplotlib.pyplot as plt
def distance(px,py):
    return math.sqrt(px*px+py*py)
def monte_carlo_pi_est(num_points):
    in_quadrant = 0
    for i in range(num_points):
        x = random.uniform(0,1)
        y = random.uniform(0,1)
        if distance(x, y) <= 1:
            in_quadrant += 1
    return (in_quadrant/num_points) * 4
pi_val=[]
for i in range (1000):
    pi_val.append(monte_carlo_pi_est(10000))
#print(pi_val)
plt.figure(figsize = (12, 8))
plt.plot(range(1000), pi_val)
plt.xlabel("iteration")
plt.ylabel("Estimation Pi")
plt.title("Change of estimated Pi over 1000 runs on 10000 darts")
plt.show()
print(sum(pi_val) / len(pi_val))
```
