```python
import random
def run_simulation():
    low = 1
    high = 100
    secret_number = random.randint(low, high)
    attempts = 0
    while low <= high:
        attempts += 1
        guess = round(low + high) // 2
        if guess == secret_number:
            print(f"Secret Number: {secret_number}")
            print(f"Total Guesses: {attempts}")
            return attempts
        elif guess > secret_number:
            print(f"Lower than {guess}")
            high = guess - 1
        else:
            print(f"Higher than {guess}")
            low = guess + 1
run_simulation()
```

```python
def bsearch(searchlist,target):
     startpos = 0
     endpos = len(searchlist) -1
     found = False
     while(not found):
          midpos = round((startpos + endpos) /2)
          print(f"Start={startpos}, End={endpos}, Mid={midpos}, Conternt={searchlist[midpos]}")
          if searchlist[midpos] == target:
               found = True
               print(f"Target found at position {midpos}")
          elif (searchlist[midpos] < target):
               startpos = midpos + 1
          elif (searchlist[midpos] > target):
                endpos = midpos - 1
     if found:
          return midpos
     else:
          return -1
my_list= [1,3,5,6,7,100,101,240,6767,8788,55555,55678]
print(bsearch(my_list, 6767))
```

```python
import random
def estimate_pi(num_samples):
    points_inside_circle = 0
    for _ in range(num_samples):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        if x**2 + y**2 <= 1:
            points_inside_circle += 1
    pi_estimate = 4 * (points_inside_circle / num_samples)
    return pi_estimate
samples = 1000000
print(f"Estimated Pi: {estimate_pi(samples)}")
```

```python
import random
import math
def distance(px, py):
    return math.sqrt(px*px + py*py)
def monte_carlo_pdl_est(num_points):
    in_puddle = 0
    for i in range(num_points):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if distance(x, y) <= 1:
            if y <x+0.5 and y>x-0.5:
                in_puddle += 1
    return in_puddle/num_points * 4
print(monte_carlo_pdl_est(100000))
```
