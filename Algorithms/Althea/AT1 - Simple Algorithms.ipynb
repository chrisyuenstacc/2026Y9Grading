```python
def bsearch(searchlist, target):
    #find out the starting and ending positions for the search
    startpos = 0
    endpos = len(searchlist) - 1
    found = False
    while (not found) and (endpos >= startpos):
        #find the midpoint of the current list to check
        midpos = round((startpos + endpos) / 2)
        print(f"Start={startpos}, End={endpos}, Mid={midpos}")
        #case 1: content in mid point == target (hit)
        if searchlist[midpos] == target:
            found = True
            print(f"Target found at position {midpos}")
        #case 2: content in mid point < target, search again
        elif (searchlist[midpos] < target):
            startpos = midpos + 1
        #case 3: content in mid point > target, search again
        elif (searchlist[midpos] > target):
            endpos = midpos - 1
        #case 4: can't split further, final check
    if found:
        return midpos
    else:
        return "NOT THERE"

mylist = [1, 3, 5, 6, 7, 100, 212, 640, 670, 6767, 67676]
print(bsearch(mylist, 2))
```
