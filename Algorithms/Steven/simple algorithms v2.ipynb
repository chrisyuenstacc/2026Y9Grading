# Simple Algorithms

```python
x = int(input("Enter a number from the following: [1, 3, 4, 32, 54, 56, 579, 777, 4324, 45353, 4394039]: "))

def bsearch(searchlist, target):
    # Find out the starting and ending position for the search
    min = 0
    max = len(searchlist) - 1
    found = False
    while (not found) and (max >= min):
        # Find the mid point of the current list to check
        middle = (min + max) // 2 # INDEX POSITION
        middle_value = searchlist[middle] # VALUE
        #print(f"min = {min}, max = {max}, middle = {middle}, searchlist = {searchlist[middle]}")
        # Case 1: Content in mid point == target, it's a HIT! Report location and finish
        if middle_value == target: 
            found = True
            #print(f"Target found at position {middle - 1}")
        # Case 2: Content in mid point < target, move to right (higher half)
        elif middle_value < target:
            min = middle + 1
        # Case 3: Content in mid point > target, move to left (lower half)
        elif middle_value > target:
            max = middle - 1
        # Case 4: Can't split further, final check
    if found:
        return middle
    else:
        return print("NUMBER NOT FOUND")
        
your_list = [1, 3, 4, 32, 54, 56, 579, 777, 4324, 45353, 4394039]

print("Target found at position", bsearch(your_list, x) + 1,  f"\nIndex position: {bsearch(your_list, x)}") # your_list, x | x is the input

# successfully loaded

```

# Monte Carlo Method - PI

```python
import random
import math
import matplotlib.pyplot as plt # pyplot is the specific function for plotting


def distance(px, py): # square
    return math.sqrt(px * px + py * py) #
    
def monte_carlo_pi_est(num_points):
    in_quadrant = 0
    for i in range(num_points):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        if distance(x,y) <= 1:
            in_quadrant += 1
    return (in_quadrant / num_points) * 4

pi_value = []

for i in range(10000):
    pi_value.append(monte_carlo_pi_est(10000))

print(pi_value)
plt.figure(figsize = (12, 8)) # inches
plt.plot(range(10000), pi_value)
plt.xlabel("Iterations")
plt.ylabel("Estimated Pi")
plt.title("Change of estimated Pi over 1000 runs on 10,000 Darts")
plt.show()


#print(monte_carlo_pi_est(100))
#print(monte_carlo_pi_est(1000))
#print(monte_carlo_pi_est(10000))
#print(monte_carlo_pi_est(100000))
#print(monte_carlo_pi_est(1000000))
#print(monte_carlo_pi_est(10000000))
#print(monte_carlo_pi_est(100000000))
#print(monte_carlo_pi_est(1000000000))
#print(monte_carlo_pi_est(10000000000))
```
