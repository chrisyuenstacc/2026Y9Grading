Battle Ship Game

```python
import random
w = 6
h = 6
ship_count = 4
shots = 15
board = {}
ship_x = random.randint(0, 6)
ship_y = random.randint(0, 6)
board[(ship_x, ship_y)] = "X"


while len(board) != 4:
    ship_x = random.randint(0, 6)
    ship_y = random.randint(0, 6)
    if (ship_x, ship_y) not in board:
        board[(ship_x, ship_y)] = "X"
    else:
        print("Regenerating...")

print(board)




while (shots !=0) and (len(board) != 0):
    guess_x = int(input("guess X"))

    guess_y = int(input("guess Y"))
    shots -= 1
    if (guess_x, guess_y) in board:
        print("hit", guess_x, guess_y, "shots left:", shots)
        del(board[(guess_x, guess_y)])
    else:
        print("miss", (guess_x, guess_y), "shots left:", shots)
    if shots == 0:
        print("lose")
    if (len(board) == 0):
        print("win")
print("ships left:",board)
```

```python
my_list = [1, 3, 4, 6, 7, 10]
def recursive_sum(list):
    if len(list) == 0:
        return 0
    return list[0] + recursive_sum(list[1:])
print(recursive_sum(my_list))
```

```python
first_numbers = [0, 1]
n = 5
def fibonacci(n):
    if n == 0:
        return 0
    elif (n == 1) or (n == 2):
        return 1
    else:
        
        return fibonacci(n-1) + fibonacci(n-2)
        
        
print(fibonacci(n))
```
