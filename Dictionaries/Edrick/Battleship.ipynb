```python
import random
where1 = []
where2 = []


def getlists():
    newlist = []
    while len(newlist) < 4:
        place = (random.randint(1,6), random.randint(1,6))
        if place not in newlist:
            newlist.append(place)
    return newlist

where1.extend(getlists())
where2.extend(getlists())
print(where1)
print(where2)

def HumanTurn():
    coord = (int(input("x?: ")), int(input("y?: ")))
    print("You entered: " + str(coord))
    if coord in where1:
        where1.remove(coord)
        print("It's a hit! " + str(len(where1)) + " ships remain")
    else:
        print("It's a miss!")
        
def BotTurn():
    coord = (random.randint(1,6), random.randint(1,6))
    print("Bot entered: " + str(coord))
    if coord in where2:
        where2.remove(coord)
        print("It's a hit! " + str(len(where2)) + " ships remain")
    else:
        print("It's a miss!")

for i in range(0,6):
    HumanTurn()
    if len(where1) == 0:
        print("You destroyed all enemies!")
        break
    BotTurn()
    if len(where2) == 0:
        print("Bot has destroyed all your ships!")
        break
    

if len(where1) < len(where2):
    print("You have more ships remaining, you win!")
elif len(where1) == len(where2):
    print("It is a draw, both players have " + len(where1) + " remaining." )
else:
    print("You have less ships remaining, you lose!")

        
```
