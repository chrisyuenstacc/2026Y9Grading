```python
PhoneBOOK = {}

while True:
    huh = input("COMMAND (1 Search | 2 Add | 3 Quit | 4 Delete | 5 View): ")

    if huh == "1":
        ask = input("name?: ")
        if ask in PhoneBOOK:
            print("Numbers:", PhoneBOOK[ask])
        else:
            print("Not Found")

    elif huh == "2":
        ask = input("name?: ")
        ask2 = input("number?: ")
        if ask not in PhoneBOOK:
            PhoneBOOK[ask] = []
        PhoneBOOK[ask].append(ask2)

        print(ask + " now has number " + ask2)

    elif huh == "3":
        print("Quitting..")
        print(PhoneBOOK)
        break

    elif huh == "4":
        ask = input("name?: ")
        if ask in PhoneBOOK:
            del(PhoneBOOK[ask])
            print(ask + " has been removed")
        else:
            print("Not Found")

    elif huh == "5":
        print(PhoneBOOK)

```
