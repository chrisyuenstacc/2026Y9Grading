```python
import random
import os
BOARD_SIZE = 5
SHIP_COUNT = 3
use = True
def clear_screen():
    os.system('cls' if os.name == 'nt' else 'clear')
def generate_computer_ships():
    ships = set()
    while len(ships) < SHIP_COUNT:
        ships.add((random.randint(0, BOARD_SIZE - 1), random.randint(0, BOARD_SIZE - 1)))
    return ships
def place_player_ships():
    player_ships = set()
    while len(player_ships) < SHIP_COUNT:
        clear_screen()
        print("DEPLOY YOUR FLEET")
        draw_board(set(), player_ships, "YOUR BOARD") 
        try:
            move = input(f"Place ship {len(player_ships) + 1} (Row Col): ").split()
            r, c = map(int, move)
            if 0 <= r < BOARD_SIZE and 0 <= c < BOARD_SIZE and (r, c) not in player_ships:
                player_ships.add((r, c))
        except:
            print("Invalid input!")
    return player_ships
def draw_board(guessed_coords, ships, title):
    print(f"\n{title}")
    print(f"    {' '.join(map(str, range(BOARD_SIZE)))}") 
    for r in range(BOARD_SIZE):
        row_display = []
        for c in range(BOARD_SIZE):
            if (r, c) in guessed_coords:
                row_display.append("X" if (r, c) in ships else "O")
            elif (r, c) in ships and title == "YOUR BOARD":
                row_display.append("S") 
            else:
                row_display.append(".")
        print(f"{r:2}  {' '.join(row_display)}")
def get_computer_guess(guessed_by_comp):
    while True:
        guess = (random.randint(0, BOARD_SIZE - 1), random.randint(0, BOARD_SIZE - 1))
        if guess not in guessed_by_comp:
            return guess
def play_game():
    own_ships = place_player_ships() 
    e_ships = generate_computer_ships()
    p_guesses = set()
    c_guesses = set()
    p_hits = 0
    c_hits = 0
    msg = "Battle Start!"
    while p_hits < SHIP_COUNT and c_hits < SHIP_COUNT:
        clear_screen()
        print(f"LAST ACTION: {msg}")
        draw_board(p_guesses, e_ships, "ENEMY BOARD")
        draw_board(c_guesses, own_ships, "YOUR BOARD")
        try:
            move = input("\nYour Move (Row Col): ").split()
            p_row, p_col = map(int, move)
            p_guess = (p_row, p_col)
            if not (0 <= p_row < BOARD_SIZE and 0 <= p_col < BOARD_SIZE) or p_guess in p_guesses:
                msg = "Invalid or duplicate move!"
                continue
            p_guesses.add(p_guess)
            if p_guess in e_ships:
                p_hits += 1
                msg = f"You HIT at {p_guess}!"
            else:
                msg = f"You MISSED at {p_guess}!"
            if p_hits == SHIP_COUNT: break
            c_guess = get_computer_guess(c_guesses)
            c_guesses.add(c_guess)
            if c_guess in own_ships:
                c_hits += 1
                msg += f" | Computer HIT you at {c_guess}!"
            else:
                msg += f" | Computer MISSED at {c_guess}!"
        except:
            msg = "Error: Enter two numbers!"
    clear_screen()
    print("GAME OVER")
    print("VICTORY!" if p_hits == SHIP_COUNT else "DEFEAT!")
if use:
    play_game()
```
