# BattleShips
6 x 6 gameboard with 4 1x1 ships and player has 6 chances

```python
import random
score = 0
ships = []
for i in range(0,4):
    numb = (random.randint(1,6), random.randint(1,6))
    ships.append(numb)    
print(ships)
for i in range(0,6):
    x = input("x coordinate")
    y = input("y coordinate")
    cords = (int(x), int(y))
    print(cords)
    for z in ships:
        if cords == z:
            score += 1
            print("hit")
    print(score)
if score == 4:
    print("you win")
else:
    print("you lose")
        
```
