# Battleship Game

```python
import random
chance = 6
x = ['1', '2', '3', '4', '5', '6']
y = ['1', '2', '3', '4', '5', '6']
cord = []

for numx in x:
    for numy in y:
        comb = numx + numy
        cord.append(comb)
print(cord)
random.shuffle(cord)
ships = cord[0:4]
print(ships)
while chance != 0:
    gunx = input("What is the x-coordinate of the point you want to fire at?")
    guny = input("What is the y-coordinate of the point you want to fire at?")
    print(f"X-coordinate: {gunx}")
    print(f"Y-coordinate: {guny}")
    for ship in ships:
        if gunx == ship[0]:
            if guny == ship[1]:
                ships.remove(ship)
                print(f"It's a hit! {len(ships)} ships remaining. \n")
                break
    else:
        print("You missed. Try again? \n")
        chance -= 1
    if ships == []:
        print("Congratulations, you win! All the battleships had been destroyed. ")
        chance = 0
        break
if ships != [] and chance == 0:
    print("You have used up all your attempts. Game over.")
```

# Two sided

```python
import random
chance = 8
numm = 0
x = ['1', '2', '3', '4', '5', '6']
y = ['1', '2', '3', '4', '5', '6']
ccord = []
hships = []
ships = []
inpu = True

for numx in x:
    for numy in y:
        comb = numx + numy
        ccord.append(comb)
random.shuffle(ccord)
cships = ccord[0:4]

while True: 
    nummx = input("What is the x-coordinate of your ship?")
    if nummx not in x:
        print("This is not a valid placement, try again. ")
    elif nummx in x:
        nummy = input("What is the y-coordinate of your ship?")
        if nummy not in y:
            print("This is not a valid placement, try again. ")
        elif nummy in y:
            numm += 1
            comb = nummx + nummy
            hships.append(comb)
    if numm == 4:
        break
print(f"Your ships are at: {hships}\n")

while chance != 0:
    print("User's turn ")
    gunx = input("What is the x-coordinate of the point you want to fire at?")
    guny = input("What is the y-coordinate of the point you want to fire at?")
    print(f"X-coordinate: {gunx}")
    print(f"Y-coordinate: {guny}")
    chance -= 1
    for cship in cships:
        if gunx == cship[0]:
            if guny == cship[1]:
                cships.remove(cship)
                print(f"It's a hit! {len(cships)} computer ships remaining. \n")
                break
    else:
        print("You missed. Try again? \n")

    print("Computer's turn ")
    random.shuffle(ccord)
    cgun = ccord[0]
    print(f"X-coordinate: {cgun[0]}")
    print(f"Y-coordinate: {cgun[1]}")
    for hship in hships:
        if cgun[0] == hship[0]:
            if cgun[1] == hship[1]:
                hships.remove(hship)
                print(f"It's a hit! {len(hships)} user ships remaining. \n")
                break
    else:
        print("It's a miss. \n")
    if cships == []:
        print("Congratulations, you win! All the battleships had been destroyed. ")
        chance = 0
        break
    elif hships == []:
        print("Game over. The computer destroyed all your battleships. ")
        break
if cships != [] and hships != [] and chance == 0:
    print("It's a tie! Both sides have used up all attempts. ")
```
