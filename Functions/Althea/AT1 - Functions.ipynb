# Silly Function #

```python
def silly(x):
    return x * -1

print(silly(100))
print(silly(6500))
```

```python
def abs_value(x):
    if x < 0:
        return x * -1
    else:
        return x

print(abs_value(-100))
print(abs_value(1000))
```

```python
def bigger(x, y):
    if x < y:
        return y
    else:
        return x

print(bigger(100,1000))
```

# Pythag #

```python
import math
#math.sqrt(x)

def pythag(a, b):
    if (a < 0) or (b < 0):
        print("This is not a valid length. Try again?")
    c = a**2 + b**2
    print(c)
    return math.sqrt(c)

print(pythag(3,4))
print(pythag(-3, -4))
```

# Fibonachi #
Find the number from the amount of recursions

```python
#Factorial - Recursion
def fib(m):
    if m == 0:
        print("Reached the base case. ")
        return 0 #Base case
    elif m == 1:
        print("Reached the base case. ")
        return 1 #Base case
    else:
        print(f"Calculating fibonachi {m}")
        return fib(m-1) + fib(m-2)

print(fib(0))
print(fib(2))
print(fib(3))
print(fib(10))
```
