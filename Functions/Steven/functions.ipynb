```python
animal_type = str(input("What is the animal type? "))
pet_name = str(input("What is the animal's name? "))

def describe_pet(animal_type = str(input("What is the animal type? ")), pet_name = str(input("What is the animal's name? "))):
    print(f"\n- I have a {animal_type}")
    print(f"- My {animal_type}'s name is {pet_name.title()}.")

describe_pet(animal_type, pet_name)

describe_pet("Siberian Husky", "Doodle")
describe_pet(pet_name = "Doodle", animal_type = "Siberian Husky")
```

# Mutable/immutable? Try this

```python
def call_by_value(x): # x = unknown variable
    x = x * 2
    print("In function value updated to:", x, ", after doubling")
    return
def call_by_reference(list):
    list.append("D")
    print("In function listed updated to", list)
    return


my_list = ["E"]
his_list = ["Z"]
num = 6 # global variable, accessable by everyone, thus 6
print("Number before =", num) # 6
call_by_value(num) # refers to inside function, thus 12
print("After function number value =", num) # 6

print("List before", my_list) # just E
print("List before", his_list) # just Z
call_by_reference(my_list)
call_by_reference(his_list)
print("After function list is ", my_list)
print("After function list is ", his_list)
```

# The Matryoshka dolls

```python
my_list = [1, 6, 103, 12, 9, 11]
sum = 0
for number in my_list:
    sum = number + sum

print(sum)
```

```python
my_list = [1, 6, 103, 12, 9, 11]

def recursive_sum(list):
    if len(list) == 0:
        return 0
    return list[0] + recursive_sum(list[1:])

recursive_sum(my_list)
```

# Factorial (!)

```python
# 6! 6 x 5 x 4 x 3 x 2 x 1 or 6 * 5!
number = 5
def factorial(n):
    if n == 2:
        return 2 # returns n and tells it
    return n * factorial(n - 1)


    
print(factorial(number))

#n = int(input("Enter a number to be factorialised (!): "))
#my_list = [1, 2, 3, 4, 5]
#print(my_list[-2])
```

# Fibonacci sequence


```python
number = 3

def fibonacci(a):
    # Bases cases n = 0, 1
    if a == 0:
        return 0 # tell the caller that it's 0
    elif (a == 1) or (a == 2):
        return 1
    elif (a == 3):
        return 2
    elif (a == 4):
        return 3
    else:
        return # tell the user that it's nothing

fibonacci(number)
```

```python
number = int(input("Enter a number to be fibonaccicised: "))
def fibonacci(a):
    if a == 0:
        return 0
    elif (a == 1) or (a == 2):
        return 1
    else:
        return fibonacci(a - 1) + fibonacci(a - 2)
print("Fibonacci:", fibonacci(number))
```
