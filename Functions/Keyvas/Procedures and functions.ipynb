```python
my_list = [1,6,107,12,9,11]
def recursion_sum(list):
    if len(list) == 0:
        return 0
    return list[0] + recursion_sum(list[1:])
print(recursion_sum(my_list))
```

#### Factorials

```python
x=6;y=1;z=1
while True:
    if y != x:
        z = z * (y + 1);y += 1
    else:
        break
print(z)
```

```python
#5! = 1*2*3*4*5 or 4!*5 and so on
num = 8
def factorial(n):
    if n == 2:
        return 2
    return n * factorial(n-1)
print(factorial(num))
```

### Fibonacci sequence

```python
num = 10
def fibonacci(n):
    if n ==0:
        return 0
    elif (n ==1) or (n==2):
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2) 
print(fibonacci(num))
```
