# Guest Management System

```python
guestlist = []
print("Guest Management System:")
num = 0
while num != -1:
    num = int(input("1: Add guest / 2: Remove guest / 3: Print guest list / -1: Quit "))
    if num == 1:
        newguest = input("Name of the guest to be added: ")
        guestlist.append(newguest)
        print(f"{newguest} added to the list.")
    elif num == 2:
        deadguest = input("Name of the guest to be removed: ")
        if deadguest in guestlist:
            guestlist.remove(deadguest)
            print(f"{deadguest} removed from the list.")
        else:
            print(f"{deadguest} not found in the list.")
    elif num == 3:
        print("Guest list: ")
        for guest in guestlist:
            print(f"{guest} will attend. ")
        print(len(guestlist))
    elif num != -1:
        print("Your number is not a valid choice. Try again? ")
```

```python
guestlist = []
print("Guest Management System:")
num = 0

def addg():
    newguest = input("Name of the guest to be added: ")
    guestlist.append(newguest)
    print(f"{newguest} added to the list.")

def remg():
    deadguest = input("Name of the guest to be removed: ")
    if deadguest in guestlist:
        guestlist.remove(deadguest)
        print(f"{deadguest} removed from the list.")
    else:
        print(f"{deadguest} not found in the list.")

def printg():
    print("Guest list: ")
    for guest in guestlist:
        print(f"{guest} will attend. ")
    print(len(guestlist))

while num != -1:
    num = int(input("1: Add guest / 2: Remove guest / 3: Print guest list / -1: Quit "))
    if num == 1:
        addg()
    elif num == 2:
        remg()
    elif num == 3:
        printg()
    elif num != -1:
        print("Your number is not a valid choice. Try again? ")
```
