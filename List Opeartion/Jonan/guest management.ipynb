```python
guest_name = []
program_use = True
def guest_add():
    name = input("Enter the name of the guest to add: ")
    guest_name.append(name)
    print(f"{name} has been added to the guest list")
def guest_remove():
    guest_remove = input("Name of the guest to be removed: ")
    if guest_remove in guest_name:
        guest_name.remove(guest_remove)
        print(f"{guest_remove} removed.")
    else:
        print("Name not found.")
def guest_list_print():
     print("Current List:", guest_name)
     print(len(guest_name) ,"people will attend")
def exit_program():
        print("Goodbye!")
while program_use:
    user_input = int(input("1: Add / 2: Remove / 3: Print / -1: Quit: "))
    if user_input == 1:
        guest_add()
    elif user_input == 2:
        guest_remove()
    elif user_input == 3:
        guest_list_print()    
    elif user_input == -1:
        exit_program()
        program_use = False
    else:
        print("Invalid option, try again.")
```
