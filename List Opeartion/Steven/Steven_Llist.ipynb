```python
# Guest Management System
 
menu = 0
guest_list = [] # creates an empty list
 
 
def add_guest():
    person = input("Enter the name of the guest to add: ")
    guest_list.append(person)
    print(f"{person} has been added to the guest list.")
 
def remove_guest():
    person = input("Guest to be removed: ")
    if person in guest_list:
        guest_list.remove(person)
        print(f"Your person {person} has been removed from the list.")
    else:
        print("The name you entered was NOT in the list.")
 
def print_list():
    print("People who are attending: ")
    for person in guest_list:
        print(f"- {person}")
    print(f"Total number of guests: {len(guest_list)}")
 
 
while menu != 4:
    menu = int(input("1: Add guest | 2: Remove guest | 3: Print guest list | 4: Quit"))
    if menu == 1:
        print("YOU ARE ABOUT TO ADD A GUEST")
        add_guest()
    elif menu == 2:
        print("YOU ARE ABOUT TO REMOVE A GUEST")
        remove_guest()
    elif menu == 3:
        print("YOU ARE ABOUT TO SEE THE CURRENT GUEST LIST")
        print_list()
    elif menu == 4:
        print("So you have finished your selections. Goodbye!")
    else:
        print("Invalid answer. Please try again.")
print(f"Total guest count: {len(guest_list)} guests")
```
